# Data Understanding

Dataset yang digunakan adalah **Iris Dataset** yang berisi informasi mengenai karakteristik bunga iris.  
Dataset ini memiliki **5 fitur utama** selain Id, yaitu panjang dan lebar sepal, panjang dan lebar petal, serta jenis spesies bunga.  

## Struktur Data

| Kolom           | Tipe Data   | Deskripsi                                                                  |
|-----------------|-------------|----------------------------------------------------------------------------|
| Id              | Numerik     | Nomor urut atau identitas unik untuk setiap data bunga.                    |
| SepalLengthCm   | Numerik     | Panjang sepal (kelopak bunga) dalam satuan sentimeter.                     |
| SepalWidthCm    | Numerik     | Lebar sepal (kelopak bunga) dalam satuan sentimeter.                       |
| PetalLengthCm   | Numerik     | Panjang petal (mahkota bunga) dalam satuan sentimeter.                     |
| PetalWidthCm    | Numerik     | Lebar petal (mahkota bunga) dalam satuan sentimeter.                       |
| Species         | Kategorikal | Jenis bunga iris (Iris-setosa, Iris-versicolor, Iris-virginica).           |

## Kesimpulan

Berdasarkan analisis awal terhadap dataset Iris, dapat ditarik beberapa poin penting:

1. Dataset terdiri dari 150 sampel dengan distribusi yang seimbang, masing-masing spesies
   (Iris-setosa, Iris-versicolor, Iris-virginica) memiliki 50 data. Hal ini memastikan
   tidak adanya masalah ketidakseimbangan kelas (class imbalance).  

2. Fitur numerik yang digunakan adalah Sepal Length, Sepal Width, Petal Length, dan Petal Width.  
   Setiap fitur memiliki rentang nilai (minâ€“max), rata-rata, dan standar deviasi yang berbeda.  

3. Fitur Petal (panjang dan lebar) menunjukkan variasi yang lebih besar dibandingkan dengan fitur Sepal.  
   Hal ini terlihat dari nilai standar deviasi yang lebih tinggi pada Petal Length dan Petal Width.  

4. Fitur Sepal cenderung lebih stabil (rentang data lebih sempit dan standar deviasi lebih kecil).  
   Sebaliknya, fitur Petal lebih bervariasi sehingga berpotensi menjadi indikator utama dalam klasifikasi spesies.  

5. Secara keseluruhan, dataset ini sangat ideal untuk digunakan sebagai studi kasus dalam analisis data
   maupun pemodelan machine learning, karena memiliki data yang bersih, seimbang, dan jelas perbedaan antar fiturnya.
